```python
import pandas as pd
```

```python
df =pd.read_csv('../Datasets_Limpios/olist_products_dataset.csv', sep=';')
df
```

```python
unicos=df['product_category_name'].unique()
len(unicos)
```

```python
rango=[]
for e in range(1,75):
    rango.append(e)
```

```python
print(rango)
```

```python
len(rango)
```

```python
du=pd.DataFrame({'category_id': rango,'product_category_name':unicos})
du
```

```python
du.to_csv('category_name_auxiliar.csv',sep=';',index=False)
```

```python
df=pd.merge(df,du, left_on='product_category_name', right_on="product_category_name", how='left')
df
```

```python
df=df.drop(columns=['product_category_name'])
df
```

```python
df.to_csv('olist_products_dataset_normalizada.csv', sep=';',index=False)
```
