```python
import pandas as pd
```

```python
df = pd.read_csv('../Tablas Preparadas/Tabla1ConEstados.csv', sep=',')
df.head()
```

```python
df1= pd.read_csv('../Tablas Preparadas/EstadosYAbrev.csv',sep=';')
df1.head()
```

```python
nueva = pd.merge(df1, df, left_on='abreviaturas',right_on='customer_state', how='left')
```

```python
nueva
```

```python
nueva.drop(columns=['abreviaturas'], inplace=True)
```

```python
nueva.drop(columns=['customer_state'],inplace=True)
```

```python
nueva.rename(columns={'Estados':'customer_state'})
```

```python
nueva.to_csv('Tabla1Final.csv', sep=';', index=False)
```
